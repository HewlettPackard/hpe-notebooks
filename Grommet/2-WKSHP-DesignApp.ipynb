## How to add a component and edit its properties

Before proceeding further, let's take a look at how to add a component and edit its properties on the Grommet designer site.

In the left sidebar, click on the **+** icon to add a new **Box** component.

![ClickPlus.png](Pictures/ClickPlus.png)

![addComponent.png](Pictures/addComponent.png)

After adding a **Box** component, select that component from the left sidebar and edit its properties on the right sidebar.

You can change the **Box** component name to **test**, direction=column, justify=between, align=center, etc.

After you finish playing with the different properties of the **Box** component that you named it as **test**, please click on the **delete** option on the top right corner to delete the selected component.

![editDeleteComponent.png](Pictures/editDeleteComponent.png)

**NOTE**

- To ignore the property value, simply mark its value as **undefined**.

- You can find the **code** , **new design from this**, **reset**, **help**,  **duplicate**, and **delete** options for each component in the top right corner, as shown below.

![ResetDelete.PNG](Pictures/resetdelete.png)

### Change Theme

Now that you have finished designing your app, let's take a look at how to apply a **theme** or change to a different one.

By default, the Grommet theme will be selected in the HPE Grommet Designer, so you will see all your buttons, links etc. in a violet color. To select a different theme, first click on your design name **student$$I**. Click on **Configure** and then click on the URL under the **Theme** section. Change it to your desired theme(hpe-next, for example) and close the pop up.

Now, you can see that the button colors have changed from violet(Grommet theme) to green, as per your theme selection (HPE).

![theme.png](Pictures/theme.png)


### Theme Mode

Grommet also offers you a couple of theme modes; **light** and **dark**.

If you don't want to view your app with too much brightness, select the dark mode and vice versa.

Click on your design name **student$$I**, Configure and select the theme mode as either light or dark to see the difference.

## Publish, Download or Generate Code

Turn your design into a real code that you can use to create a live site.

Once you finish designing your app, you can either Publish, Download, or Generate the code by clicking on the **share** icon as shown below.

![share.png](Pictures/share.png)

![share2.png](Pictures/share2.png)

First, click on **Publish** option to publish your design in grommet Designer by providing your email and a 3 digit pin. You can share the publish URL with others to share your design and get feedback. You can see my design [here.](https://designer.grommet.io/?id=StudentXX-pramod-reddy-sareddy-hpe-com)

Next, click on **Generate code** option to generate the code.

![generate-code.png](Pictures/generate-code.png)

Click on **Download** to download the design to a JSON file. You can use this as a separate backup copy, inspect and transform it with a program, or share it with someone else. You can upload it via the top left control that shows all of your designs.
