## Lab 3: Prepare the Environment

This Jupyter notebook will drive you through the following steps :

- Install nodejs, npm, and npx
- Install yarn
- Create a react app
- Adding grommet
- Add generated code from grommet designer
- Run the application
- Push the code to GitHub repo


This tutorial assumes that you don't have node.js and a package manager; either npm (npm is installed with node.js) or yarn package manager. Create react app can be installed globally or locally using npx (npx comes with npm 5.2+).

### Install nodejs, npm, and npx

- [Node Download Page](https://nodejs.org/en/download/) - The latest LTS version will have node.js, npm, and npx.

- Verify the installation

```bash
node -v
npm -v
```

![04-nodejs-install.PNG](Pictures/04-nodejs-install.png)
    
### Install yarn
    
- [Yarn Package Manager](https://yarnpkg.com/en/docs/getting-started)

- verify the installation

```bash
yarn --version
```

![05-yarn-install.PNG](Pictures/05-yarn-install.png)

### Create your app

We are going to leverage and install the awesome [create-react-app](https://facebook.github.io/create-react-app/). We will refer to it as CRA from now on.

```bash
npx create-react-app my-app-student$$I
cd my-app-student$$I
```

CRA will automatically install dependencies for you. It uses [Yarn](https://yarnpkg.com/en/) as a package manager behind the scenes. If you prefer to use NPM for your package management, you can delete the `yarn.lock` file and run `npm install` manually.

![06-yarn-start.PNG](Pictures/06-yarn-start.png)

### Adding Grommet

**NOTE**: Please follow the instructions provided on the Generate code screen

```bash
yarn add styled-components https://github.com/grommet/grommet/tarball/stable https://github.com/grommet/grommet-icons/tarball/stable https://github.com/grommet/grommet-theme-hpe/tarball/NEXT-stable
```

### Add your generated code

Replace the contents of my-app-student$$I/src/App.js with your generated code from grommet designer.

### Start the development server

```bash
yarn start
```

### Create a new repository

To put your project up on GitHub, you'll need to create a repository for it to live in.

1) After signing in to your GitHub account, in the upper-right corner of any page, use the drop-down menu, and select a New repository.

2) Type a short, memorable name for your repository. For example, "my-app-student$$I".

3) Optionally, add a description of your repository. For example, "My first repository on GitHub."

4) Choose repository visibility. For more information, see ["About repository visibility."](https://help.github.com/en/github/creating-cloning-and-archiving-repositories/about-repository-visibility)

5) Click Create repository.

### Push an existing repository from the command line

Please install git, if you don't have it already.

- [Windows](https://git-scm.com/download/win)
- [Mac](https://git-scm.com/download/mac)

Navigate to your project folder in your workstation and follow the below commands to push your code to the Git repository.

```bash
cd my-app-student$$I
git init
git add .
git commit -m "first commit"
git remote add origin https://github.com/**<your-GitHub-name>**/my-app-student$$I.git
git push -u origin master
```

**NOTE**: If you had any difficulty in designing your app, you can directly download the source code from [here.](https://github.com/reddypramod85/studentxx)

```bash
git clone https://github.com/reddypramod85/studentxx.git
cd studentxx
git remote -v
git remote set-url origin git@github.com:<YourGitHubID>/<Your_Repo_Name.git
git remote -v # verify the remote repo URL is changed to your new repo
cd src
vi App.js # modify the code to reflect your name, description and save and exit
git add .
git commit -m "first commit"
git push -u origin master
```
