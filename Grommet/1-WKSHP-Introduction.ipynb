## What is Grommet:

Grommet is a React component library that provides responsive and accessible mobile-first code components you can easily use for your projects.

Of particular note is the fact that Grommet provides powerful theming tools that let you tailor the component library to align with your color, type, component interaction, and layout needs.

Simple, consistent, and thoughtfully designed user interfaces (UI) can go a long way to integrating business applications and streamlining operations. Hewlett Packard Enterprise (HPE) is a huge proponent of simplifying complex environments. That’s why HPE developed Grommet, an open source UI development tool, and made it available to anyone looking to create consistent user experiences for their applications.

Grommet is both liked and used, sustained by a vibrant, grassroots community. It has attracted plenty of developer input with over 5,000 commits from a wide variety of different contributors.

Visit the [Grommet v2](https://v2.grommet.io) website for more information.

## Grommet Storybook

The Storybook allows you to document and explore your Grommet components. It enables you to work on your components in isolation and test them in different states.

Most importantly, Storybook enables the automatic generation of a customizable style guide site.

Grommet Storybook provides examples for each component.

Visit the [Grommet Storybook](https://storybook.grommet.io/) website for more information.

## Grommet Code Sandbox

CodeSandbox is an online editor that's built for web application development. Web application development is a growing field, and with all new configuration options, it becomes harder and harder to focus on writing code. We make this easier by doing most of the configuration for you.

This focus on web application development allows us to do many more optimizations. We can analyze npm dependencies, show custom error messages for know errors, and even make projects easily searchable by npm dependencies. The possibilities are endless!

Visit the [Grommet Code Sandbox](https://codesandbox.io/u/grommetux/sandboxes) website for more information.
